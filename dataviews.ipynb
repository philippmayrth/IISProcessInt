# This notebook is used to export csv files with different views on the data

```python
import pandas as pd
import sqlite3
```

```python
con = sqlite3.connect("../data/data.sqlite")
```

```python
SQLJoinEventWithCase = """select e.sorting, e.EVENTTIME, e.ACTIVITY_EN, e.Automation,  c.*, Revenue - Costs as "Profit"
from Pizza_Event e
inner join Pizza_Case c on e._CASE_KEY = c._CASE_KEY
order by sorting"""
df = pd.read_sql(SQLJoinEventWithCase, con)
```

```python
df
```

```python
df["VariantCategorical"] = df["Variant"].map(lambda x: f"Variant {x}") # Make it more comfortable to work with categorical data in PMTK
```

```python
df.to_csv("../data/PizzaJoinCaseAndEvent.csv")
```
