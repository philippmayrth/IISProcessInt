### Hypothesis AB20: The variant key in the attributes referes to different ways the process can be run through


#### ✅ Hypothesis AB20a: Variant 5 is not properly tracked in the ERP

Accept: There is sufficient evidence (no available event data) to conclude that this variant is not being tracked in the ERP

```python
import pandas as pd
import sqlite3

conn = sqlite3.connect("../data/data.sqlite")
AB20SQL = """select count(*)
from Pizza_Event e
left join Pizza_Case c on e._CASE_KEY = c._CASE_KEY
where Variant = 5
order by sorting"""
dfAB20 = pd.read_sql_query(AB20SQL, conn)
print("left Join between event and case table returns 0 results:")
print()
print(AB20SQL)
print(dfAB20)
print()
print()

AB20SQL = """select count(*) from pizza_event where _CASE_KEY in (select _CASE_KEY from pizza_case where variant = 5)"""
dfAB20 = pd.read_sql_query(AB20SQL, conn)
print("so does a subselect:")
print()
print(AB20SQL)
print(dfAB20)

del dfAB20
del AB20SQL
```
